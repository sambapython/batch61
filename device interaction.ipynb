```python
from subprocess import check_output
data = check_output(["ps","-aux"])
for row in data.splitlines():
    row = row.decode("utf-8")
    if "sublime" in row:
        pid = row.split()[1]
        cmd = "kill -9 %s"% pid
        os.system(cmd)
        break
```
